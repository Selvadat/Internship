```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\ajay\Desktop\T3.csv")
df
```

```python
# Average Electrical generation capacity(capacity_mw) according to the primary_fuel type
df.groupby(['primary_fuel']).agg({'capacity_mw':'mean'})
```

```python
# Creating a new Dataframe by removing the null values in latitude and longitude
mes=df.loc[df['latitude'].notna()==True]
```

```python
import warnings
warnings.filterwarnings('ignore')
# Prediction Primary Fuel according to Capacity_mw
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
y=OrdinalEncoder().fit_transform(mes['primary_fuel'].to_numpy()[:,np.newaxis]).ravel()
X=mes.loc[:,['capacity_mw','latitude','longitude']]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
for estimators in [SVC(class_weight='balanced'),DecisionTreeClassifier(class_weight='balanced'),KNeighborsClassifier(),RandomForestClassifier(class_weight='balanced')]:
    model=estimators
    #Identifying the parameters of an algorithm
    model.get_params()
   # Passing the model parameters to train the dataset
    model.fit(X_train,y_train)
   # Validating the model using test dataset
    y_pred=model.predict(X_test)
    # Finding the Accuracy of the model
    print(estimators,':',accuracy_score(y_test,y_pred))
   # Classification Report 
    print(classification_report(y_test,y_pred))
  # Predicting power
    print(model.predict(mes.loc[:,['capacity_mw','latitude','longitude']]))

# Even if the accuracy score is less SVC estimator predicting the right 
```
