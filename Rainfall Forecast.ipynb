```python
alb.dropna(inplace=True) # Removing null values in the dataset
```

```python
alb
```

```python
from math import sin,pi
days_in_year=365
def encode_day(timestamp):
            return round(abs(sin(pi*timestamp.timetuple().tm_yday/days_in_year)),5)        
def is_weekend(timestamp):
            if timestamp.weekday() > 4:
                return 1
            return 0       
        
# Converting the date into an encoded format using the above formulas
```

```python

alb['encode_day']=[encode_day(t) for t in alb['Date']]
alb['is_weekend']=[is_weekend(t) for t in alb['Date']]
alb
```

```python
tmp_alb=df.groupby(['Location']).get_group('Albury').loc[:,['Date','MinTemp','MaxTemp']].set_index('Date')
```

```python
val=[]
for i in alb['Date']:
    j=i-pd.Timedelta('1days')
    l=i-pd.Timedelta('7days')
    val.append(dict(tmp_alb.loc[l:j,:].mean()))

avg_tmp=pd.DataFrame(val)
alb['Avg_MinTmp_7']=avg_tmp['MinTemp'].values
alb['Avg_MaxTmp_7']=avg_tmp['MaxTemp'].values
alb
```

```python
alb.drop(['Date'],axis=1,inplace=True)
```

```python
alb=alb.round(3)
```

```python
alb['Avg_MinTmp_7']=alb['Avg_MinTmp_7'].fillna(9.876312)
alb['Avg_MaxTmp_7']=alb['Avg_MaxTmp_7'].fillna(22.726626)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.ensemble import RandomForestRegressor
X=alb.drop(['Rainfall','Avg_MinTmp_7','Avg_MaxTmp_7','Rainfall_7d','Rainfall_6d','Rainfall_5d','Rainfall_4d','Rainfall_3d','Rainfall_2d'],axis=1)
y=alb['Rainfall']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
reg=RandomForestRegressor()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
print(r2_score(y_test,y_pred))
mean_squared_error(y_test,y_pred)
```

```python
# Time Series
ts=df.groupby(['Location']).get_group('Albury').loc[:,['Date','Rainfall']].set_index('Date')
ts=ts.resample('1D').sum()
ts
```

```python
ts.plot(figsize=(10,8))
```

```python
from statsmodels.tsa.stattools import adfuller
adfuller_result=adfuller(ts['Rainfall'].values,autolag='AIC')
print('AdfStatistics:',adfuller_result[0])
print('P-values:',adfuller_result[1])
# Note the P-value is less than 0.05
# Reject the Null Hypothesis Means The time series data is Stationary
```

```python
# TimeSeries Model using Fbprophet
fb=ts.reset_index()
fb.columns=['ds','y']
from fbprophet import Prophet
model=Prophet(daily_seasonality=True)
model.fit(fb)
future=model.make_future_dataframe(periods=30)
forecast=model.predict(future)
forecast.tail(30)
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,LabelBinarizer
df['Location']=OrdinalEncoder().fit_transform(df['Location'].to_numpy()[:,np.newaxis])
df['MinTemp']=df['MinTemp'].fillna(df['MinTemp'].mean())
df['MaxTemp']=df['MaxTemp'].fillna(df['MaxTemp'].mean())
df
```

```python
df['Rainfall']=df['Rainfall'].fillna(0)
```

```python
df['RainToday']=df['RainToday'].fillna('No')
```

```python
df['RainTomorrow']=df['RainTomorrow'].fillna('No')
```

```python
df.info()
```

```python
df['WindGustSpeed']=df['WindGustSpeed'].fillna(df['WindGustSpeed'].mean())
df['WindSpeed9am']=df['WindSpeed9am'].fillna(df['WindSpeed9am'].mean())
df['WindSpeed3pm']=df['WindSpeed3pm'].fillna(df['WindSpeed3pm'].mean())
df['Humidity9am']=df['Humidity9am'].fillna(df['Humidity9am'].mean())
df['Humidity3pm']=df['Humidity3pm'].fillna(df['Humidity3pm'].mean())
df['Pressure9am']=df['Pressure9am'].fillna(df['Pressure9am'].mean())
df['Pressure3pm']=df['Pressure3pm'].fillna(df['Pressure3pm'].mean())
df['Cloud9am']=df['Cloud9am'].fillna(df['Cloud9am'].mean())
df['Cloud3pm']=df['Cloud3pm'].fillna(df['Cloud3pm'].mean())
df['Temp9am']=df['Temp9am'].fillna(df['Temp9am'].mean())
df['Temp3pm']=df['Temp3pm'].fillna(df['Temp3pm'].mean())
df.info()
```

```python
df['RainToday']=LabelBinarizer().fit_transform(df['RainToday'])
df['RainTomorrow']=LabelBinarizer().fit_transform(df['RainTomorrow'])
df
```

```python
df.columns
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,roc_auc_score,classification_report
X_b=df.loc[:,['Location', 'MinTemp', 'MaxTemp', 'Rainfall','WindGustSpeed','WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm', 'RainToday']]
y_b=df['RainTomorrow']
X_train,X_test,y_train,y_test=train_test_split(X_b,y_b,test_size=0.3,random_state=42)
clf=BernoulliNB()
clf.fit(X_train,y_train)
pred=clf.predict(X_test)
accuracy_score(y_test,pred)
```

```python
roc_auc_score(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

```python
clf_1=RandomForestClassifier()
clf_1.fit(X_train,y_train)
pred_1=clf_1.predict(X_test)
accuracy_score(y_test,pred_1)
```

```python
roc_auc_score(y_test,pred_1)
```

```python
print(classification_report(y_test,pred_1))
```
