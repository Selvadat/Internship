```python
df['Dependents']=df['Dependents'].str.replace('3+','3',regex=False)
```

```python
df.groupby(['Self_Employed'],dropna=False).agg({'ApplicantIncome':"mean"}).sort_values(by='ApplicantIncome',ascending=False)
```

```python
df['Income']=df['ApplicantIncome']+df['CoapplicantIncome']
df
```

```python
df['Ratio_Loan_Income']=df['LoanAmount']/df['Income']
df
```

```python
df.groupby(['Loan_Status']).agg({'Ratio_Loan_Income':"mean"}).round(3)
```

```python
df['Ratio_Loan_Income']=df['Ratio_Loan_Income'].fillna(np.mean(df['Ratio_Loan_Income'])).round(3)
df
```

```python
#Handling Missing Values in Loan Amount Column
df['LoanAmount']=(df['Income']*df['Ratio_Loan_Income']).round(1)
df
```

```python
df.loc[df['Loan_Amount_Term'].isna()==True]
```

```python
df.info()
```

```python
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(360.0)
df
```

```python
df.groupby(['Self_Employed'],dropna=False).agg({'Income':'mean','ApplicantIncome':"mean"})
```

```python
f=[]
for j in df.loc[df['Self_Employed'].isna()==True,'ApplicantIncome']:
    if j>5000:
        f.append('Yes')
    else:
        f.append('No')
        
f
```

```python
df.loc[df['Self_Employed'].isna()==True,'Self_Employed']=f
```

```python
df['Credit_History']=df['Credit_History'].fillna(1.0)
```

```python
df['Dependents']=df['Dependents'].fillna(0)
```

```python
df['Married']=df['Married'].fillna('Yes')
```

```python
df['Gender']=df['Gender'].fillna('Male')
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelBinarizer,OrdinalEncoder
df['Loan_Status']=LabelBinarizer().fit_transform(df['Loan_Status']) # Yes:1 No:0
df['Gender']=LabelBinarizer().fit_transform(df['Gender']) # Male:1 Female:0
df['Self_Employed']=LabelBinarizer().fit_transform(df['Self_Employed']) #  No: 0 , Yes:1
df['Property_Area']=OrdinalEncoder().fit_transform(df['Property_Area'].to_numpy()[:,np.newaxis]) # Rural:0 SemiUrban:1 Urban:2
df['Education']=LabelBinarizer().fit_transform(df['Education']) # Graduate:0 Non-Graduate:1
df['Married']=LabelBinarizer().fit_transform(df['Married'])# No:0 Yes:1
```

```python
df['Dependents']=df['Dependents'].astype('int')
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,roc_auc_score
X=df.loc[:,['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']]
y=df['Loan_Status']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
log=LogisticRegression(solver='liblinear')
log.fit(X_train,y_train)
y_pred=log.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
roc_auc_score(y_test,y_pred)
# Roc_auc_score is greater than 50 
# Logistic Regression is a good model
```
