```python
import pandas as pd
import numpy as np
df=pd.read_excel(r"C:\Users\selva\Downloads\Data_Train.xlsx")
df
```

```python
test_df=pd.read_excel(r"C:\Users\ajay\Downloads\test_set.xlsx")
test_df
```

```python
#Combining train and test data as a Single dataframe
data=pd.concat([df,pd.concat([test_df,pd.read_excel(r"C:\Users\ajay\Downloads\Sample_submission.xlsx")],axis=1)])
```

```python
# Checking the datatypes of a dataframe
data.dtypes
```

```python
# Transforming the datatypes as per our Analysis (EDA)
data['Arrival_Time']=pd.to_datetime(data['Arrival_Time']).dt.time
```

```python
# Converting the Duration time in hours
data['Duration']=pd.to_timedelta(data['Duration']).dt.seconds/3600
```

```python
# Droping na values in a dataset
data.dropna(inplace=True)
```

```python
data['Total_Stops']=data['Total_Stops'].str.split('stop',expand=True)[0].str.replace('non-','0',regex=False)
data
```

```python
data['Total_Stops']=data['Total_Stops'].astype('int')
data
```

```python
data['Airline'].unique()
```

```python
#Generating a Report where The average Price of the flight between two stations depends upon the Airline
data.loc[(data['Source']=='Banglore') & (data['Destination']=='New Delhi')].groupby(['Airline','Total_Stops']).agg({'Price':"mean"}).unstack(0,fill_value=0).round(2)
```

```python
data.loc[(data['Source']=='Kolkata') & (data['Destination']=='Banglore')].groupby(['Airline','Total_Stops']).agg({'Price':"mean"}).unstack(0,fill_value=0).round(2)
```

```python
data=data.round(2)
data
```

```python
# Changing the Price into 1000's
data['Price']=data["Price"]/1000
```

```python
from sklearn.preprocessing import OneHotEncoder
tr=pd.DataFrame(np.hstack((OneHotEncoder().fit_transform(data['Airline'].to_numpy()[:,np.newaxis]).toarray(),data.loc[:,['Duration','Total_Stops','Price']])))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
X=tr.iloc[:,:-1]
y=tr.iloc[:,-1].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
reg=RandomForestRegressor(n_estimators=200,criterion='mae')
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
mean_absolute_percentage_error(y_test,y_pred)
```
